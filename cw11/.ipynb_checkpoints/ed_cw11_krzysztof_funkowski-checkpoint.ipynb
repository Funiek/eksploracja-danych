```python
setosa = df[df['Species'].str.contains('Iris-setosa')]
versicolor = df[df['Species'].str.contains('Iris-versicolor')]
virginica = df[df['Species'].str.contains('Iris-virginica')]
```

### Implementacja algorytmu K-means

```python
def k_means(data, k, max_iters=100, centroids = None, labels = None):
    
    # Inicjalizacja centroidów wybierając losowe, jeżeli nie były podane jako parametr funkcji
    if centroids is None:
        centroids = data[np.random.choice(data.shape[0], k, replace=False)]

    # Tabela na historię centroidów
    centroids_history = [centroids.copy()]
    
    for _ in range(max_iters):
        # Jeżeli nie zostały w parametrze podane etykietowania wstępne
        if labels is None:
            # Przypisanie punktów do najbliższych centroidów
            # data - centroids[:, np.newaxis - Dla każdego punktu danych odjęta zostaje wartość odpowiadająca współrzędnym centroidów
            # np.linalg.norm - normalizacja wektora
            norm = np.linalg.norm(data - centroids[:, np.newaxis], axis=2)

            # wyznaczenie indeksów dla każdego punktu
            labels = np.argmin(norm, axis=0)
        
        # Aktualizacja centroidów
        # Dla każdego klastra obliczane są średnie wartości dla każdej kolumny
        new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])
        
        # Sprawdzenie warunku czy nowe centroidy są 'lepsze' i jeśli nie to wyjście z pętli.
        if np.all(centroids == new_centroids):
            break

        # Wstawienie nowych centroidów jako warunku odniesienia
        centroids = new_centroids.copy()

        centroids_history.append(new_centroids.copy())
    
    return centroids, labels, centroids_history
```

### Ustawienia uruchamiania

```python
k = 3
max_iterations = 100
```

### Wizualizacja zbioru

```python
plt.scatter(setosa['SepalLengthCm'], setosa['SepalWidthCm'], color='red', marker='o', label='Iris-setosa')
plt.scatter(versicolor['SepalLengthCm'], versicolor['SepalWidthCm'], color='green', marker='o', label='Iris-versicolor')
plt.scatter(virginica['SepalLengthCm'], virginica['SepalWidthCm'], color='blue', marker='o', label='Iris-virginica')
plt.xlim(4,8)
plt.ylim(1,5)
plt.legend()
plt.title('Zbiór Iris')
plt.show()
```

### Uruchomienie algorytmu K-means

```python
centroids, labels, centroids_history = k_means(dataset, k, max_iterations)
```

### Tabele na rozłożenie danych o historii centroidów dla każdej klasy

```python
setosa_history = []
versicolor_history = []
virginica_history = []
```

### Przetworzenie danych historii położenia centroidów do późniejszego zobrazowania na wykresie

```python
for epoch in centroids_history:
    for idx, species_history in enumerate(epoch):
        if idx % 3 == 0:
            setosa_history.append([species_history[0], species_history[1]])
        elif idx % 3 == 1:
            versicolor_history.append([species_history[0], species_history[1]])
        elif idx % 3 == 2:
            virginica_history.append([species_history[0], species_history[1]])

setosa_history = np.array(setosa_history)
versicolor_history = np.array(versicolor_history)
virginica_history = np.array(virginica_history)
```

### Wykres wyników

```python
plt.scatter(setosa['SepalLengthCm'], setosa['SepalWidthCm'], color='red', marker='o', label='Iris-setosa')
plt.plot(setosa_history[:, 0], setosa_history[:, 1],  color='cyan', label='historia Iris-setosa')

plt.scatter(versicolor['SepalLengthCm'], versicolor['SepalWidthCm'], color='green', marker='o', label='Iris-versicolor')
plt.plot(versicolor_history[:, 0], versicolor_history[:, 1],  color='cyan', label='historia Iris-versicolor')

plt.scatter(virginica['SepalLengthCm'], virginica['SepalWidthCm'], color='blue', marker='o', label='Iris-virginica')
plt.plot(virginica_history[:, 0], virginica_history[:, 1],  color='cyan', label='historia Iris-virginica')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='cyan', label='centroidy')
plt.title('Znalezione centroidy algorytmem K-means')
plt.xlim(4,8)
plt.ylim(1,5)
plt.legend()
plt.show()
```

### Wstępny podział na grupy

```python
initial_labels = np.zeros(data.shape[0])
initial_labels[dataset[:, 0] > 5.0] = 1
initial_labels[dataset[:, 0] > 6.5] = 2
initial_labels
```

```python
plt.scatter(dataset[np.where(initial_labels == 0)[0]][:,0], dataset[np.where(initial_labels == 0)[0]][:,1], color='red', marker='o', label='Iris-setosa')
plt.scatter(dataset[np.where(initial_labels == 1)[0]][:,0], dataset[np.where(initial_labels == 1)[0]][:,1], color='green', marker='o', label='Iris-versicolor')
plt.scatter(dataset[np.where(initial_labels == 2)[0]][:,0], dataset[np.where(initial_labels == 2)[0]][:,1], color='blue', marker='o', label='Iris-virginica')
plt.xlim(4,8)
plt.ylim(1,5)
plt.legend()
plt.title('Zbiór Iris')
plt.show()
```

```python
centroids, labels, centroids_history = k_means(dataset, k, max_iterations, labels=initial_labels)

setosa_history = []
versicolor_history = []
virginica_history = []

for epoch in centroids_history:
    for idx, species_history in enumerate(epoch):
        if idx % 3 == 0:
            setosa_history.append([species_history[0], species_history[1]])
        elif idx % 3 == 1:
            versicolor_history.append([species_history[0], species_history[1]])
        elif idx % 3 == 2:
            virginica_history.append([species_history[0], species_history[1]])

setosa_history = np.array(setosa_history)
versicolor_history = np.array(versicolor_history)
virginica_history = np.array(virginica_history)

plt.scatter(setosa['SepalLengthCm'], setosa['SepalWidthCm'], color='red', marker='o', label='Iris-setosa')
plt.plot(setosa_history[:, 0], setosa_history[:, 1],  color='cyan', label='historia Iris-setosa')

plt.scatter(versicolor['SepalLengthCm'], versicolor['SepalWidthCm'], color='green', marker='o', label='Iris-versicolor')
plt.plot(versicolor_history[:, 0], versicolor_history[:, 1],  color='cyan', label='historia Iris-versicolor')

plt.scatter(virginica['SepalLengthCm'], virginica['SepalWidthCm'], color='blue', marker='o', label='Iris-virginica')
plt.plot(virginica_history[:, 0], virginica_history[:, 1],  color='cyan', label='historia Iris-virginica')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='cyan', label='centroidy')
plt.title('Znalezione centroidy algorytmem K-means')
plt.xlim(4,8)
plt.ylim(1,5)
plt.legend()
plt.show()
```
