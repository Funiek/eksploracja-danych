```python
def generate_dataset(N):
    dataset = []
    for _ in range(N):
        dataset.append([random.random(), random.random()])
        
    return dataset
```

```python
dataset = np.array(generate_dataset(500))
```

```python
def rotate(phi):
    rad = np.radians(phi)
    return np.array([[np.cos(rad), np.sin(rad)],
                     [-np.sin(rad), np.cos(rad)]])
```

```python
def stretch(x, y, z=None):
    if z is None:
        return np.array([[x, 0],
                         [0, y]])
    else:
        return np.array([[x, 0, 0],
                         [0, y, 0],
                         [0, 0, z]])
```

```python
def perform_transformation(dataset, rotation_angle, x, y, z=None):
    result_dataset = copy.deepcopy(dataset)
    
    rotation = rotate(rotation_angle)
    result_dataset = np.dot(result_dataset, rotation)
    
    stretching = stretch(x, y, z)
    result_dataset = np.dot(result_dataset, stretching)
    
    return result_dataset
```
